### 4.4 Time Analysis

#### revenue month analysis

```python
monthly_revenue = (
    df.groupby('Month')['Revenue']
      .sum()
      .reset_index()
)

monthly_revenue
```

```python
month_order = [
    'January','February','March','April','May','June',
    'July','August','September','October','November','December'
]

monthly_revenue['Month'] = pd.Categorical(
    monthly_revenue['Month'],
    categories=month_order,
    ordered=True
)

monthly_revenue = monthly_revenue.sort_values('Month')
```

#### revenue day analysis

```python
daily_revenue = (
    df.groupby('Day')['Revenue']
      .sum()
      .reset_index()
)

daily_revenue
```

```python
day_order = [
    'Monday',
    'Tuesday',
    'Wednesday',
    'Thursday',
    'Friday',
    'Saturday',
    'Sunday'
]

daily_revenue['Day'] = pd.Categorical(
    daily_revenue['Day'],
    categories=day_order,
    ordered=True
)

daily_revenue = daily_revenue.sort_values('Day')
```

#### peak hour

```python
hourly_sales = (
    df.groupby('Hour')['Revenue']
      .sum()
      .reset_index()
)

hourly_sales
```

### 4.5 spatial analysis

```python
store_location = pd.DataFrame({
    'store_location':[
        "Hell's Kitchen",
        "Astoria",
        "Lower Manhattan"
    ],
    'Latitude':[
        40.764,
        40.764,
        40.707
    ],
    'Longitude':[
        -73.991,
        -73.923,
        -74.011
    ]
})
```

```python
df = df.merge(
    store_location,
    on='store_location',
    how='left'
)
```

```python
store_revenue = (
    df.groupby(
        ['store_location','Latitude','Longitude']
    )['Revenue']
    .sum()
    .reset_index()
)

store_revenue
```

```python
fig = px.scatter_mapbox(
    store_revenue,
    lat='Latitude',
    lon='Longitude',
    size='Revenue',
    color='Revenue',
    hover_name='store_location',
    zoom=10,
    mapbox_style='open-street-map',
    title='Revenue by Store Location'
)

fig.show()
```
