```python
import pandas as pd

df = pd.read_excel("coffee Shop Sales.xlsx")
```

SPATIAL ANALYSIS

```python
import plotly.express as px
store_map = (
    df.groupby(['store_location','Latitude','Longitude'])['Revenue']
      .sum()
      .reset_index()
)

fig = px.scatter_mapbox(
    store_map,
    lat="Latitude",
    lon="Longitude",
    size="Revenue",
    color="Revenue",
    hover_name="store_location",
    mapbox_style="open-street-map"
)

fig.show()
```
